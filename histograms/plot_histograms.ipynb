```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img_pre = cv.imread('pre.png', 0)
plt.hist(img_pre.ravel(), 256, [0,256])
plt.show()
```

```python
img_post = cv.imread('post.png', 0)
plt.hist(img_post.ravel(), 256, [0,256])
plt.show()
```

```python
# Plot together
hist_pre = cv.calcHist([img_pre],[0],None,[256],[0,256])
hist_post = cv.calcHist([img_post],[0],None,[256],[0,256])

plt.plot(hist_pre)
plt.plot(hist_post)
plt.savefig('hist_pre_post.png')
plt.show()
```
